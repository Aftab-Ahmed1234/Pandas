# Data Cleaning

```python
# Bad data could be:

# Empty cells
# Data in wrong format
# Wrong data
# Duplicates
```

```python
#DataSet with Missing Values
import pandas as pd
import numpy as np
data={
    "Name":["Aftab",None,"Arslan","Abdullah"],
    "Age":[20,18,np.nan,21],
    "Marks":[20,np.nan,85,95]
    
}
df=pd.DataFrame(data)
print(df)
```

```python
# Checkin Missing Values
print(df.isnull())
```

```python
# Count Missig values in each column
print(df.isnull().sum())
```

```python
# Remove Missing Values
clean_df=df.dropna()
print(clean_df.to_string())
```
